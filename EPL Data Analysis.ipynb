# Find the number of games each referee officiated

```python
Referee = {}
for row in mainData:
    referee = row[8]
    if referee not in Referee:
        Referee[referee] = 1
    else:
        Referee[referee] += 1
print(Referee)
```

```python
Referee_data = pd.DataFrame.from_dict(Referee,orient='index')
Referee_data 
```
